Установим нужные библиотеки

```python
import requests as rq
from bs4 import BeautifulSoup, NavigableString
```

Создаем функцию, которая собирает нужные ссылки с сайта для дальнейшего парсинга

```python
def parse_page_with_articles(url): 
    page = rq.get(url)
    soup = BeautifulSoup(page.text, features="html.parser")

    urls = []

    for link in soup.find_all('a'):
        urls.append(link.get('href'))

    releases = []

    for link in urls:
        if link is None:
            continue
        if 'public' in link:
            releases.append(link)

    realeases_links = []
    page_links = []
    bad_links = ['/public/', '/public/?popular']
    for link in releases:
        if 'PAGEN_1' in link:
            page_links.append('https://psychologyjournal.ru' + link)

        elif link not in bad_links:
            realeases_links.append('https://psychologyjournal.ru' + link)

    return realeases_links, page_links
```

Функция для того, чтобы достать нужные данные со страницы

```python
def GetArticle(url): 
    page0 = rq.get(url)
    soup0 = BeautifulSoup(page0.text, features="html.parser")
    title = soup0.find_all('meta', {'property': 'og:title'})[0].attrs['content']

    tag_with_text = soup0.find_all('div', {'id': 'page_detail_text'})[0]
    new_text = ""
    for t in tag_with_text.contents:
        if isinstance(t, NavigableString):
            new_text += str(t).strip()

    final_text = new_text.strip().replace('\xa0', ' ')
    return url, title, final_text
```

Парсим страницы, берем только часть статей

```python
pages = ['https://psychologyjournal.ru/public/']
looked_pages = []
realeases_links = []

counter = 0

while pages:  # while len(pages) != 0:
    page = pages.pop(0)  # Берем первую ссылку со статьями
    print('Парсим страницу', page)
    # парсим все ссылки, разделяя на 2 списка: список с ссылками на статьи и список с ссылками на страницы
    articles_urls, parsed_page_urls = parse_page_with_articles(page)
    realeases_links.extend(articles_urls)

    looked_pages.append(page)
    for p_url in parsed_page_urls:
        if p_url not in looked_pages and p_url not in pages:
            pages.append(p_url)
            counter += 1
            if counter == :
              break
```

```python
counter = 0
articles = []
for link in realeases_links:
  try:
    new = GetArticle(link)
    articles.append(new)
    print(new)
    counter += 1
  except:
    pass
```

Собираем данные в датафрейм

```python
import pandas as pd 

df = pd.DataFrame(articles)
df.head()

df.columns = ['link', 'title', 'text']

df.head(12)

df.to_csv('Корпус Вестник психологии.csv')
```
